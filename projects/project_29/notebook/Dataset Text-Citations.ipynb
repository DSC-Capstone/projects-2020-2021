```python
import wikipedia
articles_1 = wikipedia.search("covid")
articles_2 = wikipedia.search("coronavirus")
articles_2
```

```python
total = articles_1 + list(set(articles_2) - set(articles_1))
total
```

```python
columns = ['Article Name', 'Text', 'Citations']
num = 0 
df = pd.DataFrame(columns=columns)
df['Article Name'] = total
```

```python
df
```

```python
d = dict()
c = dict()
```

```python
for i in total: 
    print(i)
    
    if(i == 'COVID-19 vaccine' or i == 'List of deaths due to COVID-19'): 
        None; 
    else: 
        text = wikipedia.page(i).content
        citations =  wikipedia.page(i).references 
        d[i] = text
        c[i] = citations
```

```python
df['Text']= df['Article Name'].map(d)
```

```python
df['Citations']= df['Article Name'].map(c)
```
