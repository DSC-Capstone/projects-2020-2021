```python
def convert_artist(name):
    return sp.search(name, type='artist')['artists']['items'][0]['id']

new_songs = []
for artist in rec_artists:
    for track in sp.artist_top_tracks(convert_artist(artist))['tracks']:
        new_songs.append(track['id'])
new_songs = list(set(new_songs)-set(seed_tracks))
new_songs
```

```python
audio_features = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 
                  'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
```

```python
dfrow = []
for song in sp.audio_features(seed_tracks):
    row = [song['id']]
    for feature in audio_features:
        row.append(song[feature])
    dfrow.append(pd.DataFrame([row]))
old_df = pd.concat(dfrow).reset_index(drop=True)
old_df
```

```python
dfrow = []
for song in sp.audio_features(new_songs):
    row = [song['id']]
    for feature in audio_features:
        row.append(song[feature])
    dfrow.append(pd.DataFrame([row]))
new_df = pd.concat(dfrow).reset_index(drop=True)
new_df
```

```python
preference = old_df.mean()
preference
```

```python
new_df.mean()
```

```python
distances = []
for index, row in new_df.iterrows():
    euc = 0
    for i in range(1,12):
        euc += (row[i]-preference[i])**2
    distances.append(round(np.sqrt(euc),1))
new_df['distance'] = distances
new_df
```

```python
new_df.sort_values(by='distance')[:30]
```

```python
recommendations = new_df.sort_values(by='distance')[0][:30].to_list()
recommendations
```
