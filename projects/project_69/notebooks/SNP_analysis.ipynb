```python
import pandas as pd
```

```python
file = open("../data/snp/gid.txt","r") 
Counter = 0

# Reading from file 
Content = file.read() 
CoList = Content.split("\n") 
  
for i in CoList: 
    if i: 
        Counter += 1
          
```

```python
chro = ['Chromosome'] * Counter
pos = [0] * Counter
geneid = [[]] * Counter
index = 0
with open('../data/snp/gid.txt') as f:
    for line in f:
        lst = line.split('\t')
        pos[index] = lst[1]
        geneid[index] = lst[3:]
        geneid[index][-1] = geneid[index][-1][:-1]
        index += 1
```

```python
gene = [[]] * Counter
index= 0
with open('../data/snp/gene.txt') as f:
    for line in f:
        lst = line.split('\t')
        gene[index] = lst[3:]
        gene[index][-1] = gene[index][-1][:-1]
        index += 1
```

```python
ale = [[]] * Counter
index= 0
with open('../data/snp/ale.txt') as f:
    for line in f:
        lst = line.split('\t')
        ale[index] = lst[2:]
        ale[index][-1] = ale[index][-1][:-1]
        index += 1
```

```python
eff = [[]] * Counter
index= 0
with open('../data/snp/eff.txt') as f:
    for line in f:
        lst = line.split('\t')
        eff[index] = lst[2:]
        eff[index][-1] = eff[index][-1][:-1]
        index += 1
```

```python
ft = [[]] * Counter
index= 0
with open('../data/snp/ft.txt') as f:
    for line in f:
        lst = line.split('\t')
        ft[index] = lst[3:]
        ft[index][-1] = ft[index][-1][:-1]
        index += 1
```

```python
base = pd.read_csv('../data/snp/ba.txt', sep = '\t')
base
```

```python
base['base_c'] = base.REF + ' > ' + base.ALT.str[0]
bc = base.groupby('base_c').agg('count')['CHROM'].sort_values(ascending = False)
bc
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
cmap = mcolors.LinearSegmentedColormap.from_list("", [ "skyblue" , "midnightblue"])
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(bc.index,bc/bc.sum(),color=cmap(bc/bc.sum()/(bc/bc.sum()).max()))
```

```python
gid = pd.DataFrame(columns = ['Chrom','POS','GENEID'])
gid['Chrom'] = chro[1:]
gid['POS'] = pos[1:]
gid['GENEID'] = geneid[1:]
```

```python
gid['GENE'] = gene[1:]
gid['ALLELE'] = ale[1:]
gid['EFF'] = eff[1:]
gid['FEATURE'] = ft[1:]
gid['BASE'] = base.REF + ' > ' + base.ALT.str[0]
```
