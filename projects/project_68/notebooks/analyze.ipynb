```python
alz_df = filtered_names.loc[:,filtered_names.columns.isin(coldata_names.groupby('GROUP')['Run'].apply(list).iloc[0])]
```

# ENST00000390223.3

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure()

row2 = alz_df[alz_df['target_id'] == 'ENST00000390223.3']
row2 = row2.set_index(['target_id']).T
# sns.boxplot(row)
# plt.boxplot(row2)


row = control_df[control_df['target_id'] == 'ENST00000390223.3']
row = row.set_index(['target_id']).T
# sns.boxplot(row)
plt.boxplot([row2['ENST00000390223.3'].values,row['ENST00000390223.3'].values])


xaxis = ["Alzheimer's","Control"]
plt.xticks([1,2],xaxis)
# plt.xlim(-2, len(ticks)*2)
plt.title("Most Upregulated of DESeq2 (ENST00000390223.3)")
plt.grid()
plt.ylim(-1, 5)
plt.ylabel("counts")
```
