```python
import pandas as pd
import os
import sys
from IPython.core.display import Image
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import os
from scipy.signal import find_peaks
from sklearn.metrics import accuracy_score
import numpy
import sklearn.model_selection as model_selection
```

```python
raw_path = "../data/in/"
entries = os.listdir(raw_path)
entries
```

```python
lab = []
for i in entries:
    if "live" in i:
        lab.append(1)
    if "streaming" in i:
        lab.append(0)
lab
```

# Feature section

```python
feat1 = []
for i in entries:
    temp = raw_path + i
    file = pd.read_csv(temp, index_col=0).reset_index()
    group_file = file.groupby("Time").sum().reset_index()
    vpr = (len(group_file)-1)/(group_file["Time"][len(group_file)-1]-group_file["Time"][0])
    feat1.append(vpr)
feat1
```

```python
"np.mean(i['2->1Bytes'])"
feat2 = []
for i in entries:
    temp = raw_path + i
    file = pd.read_csv(temp, index_col=0).reset_index()
    peaks, _ = find_peaks(file['2->1Bytes'], height=np.mean(file['2->1Bytes']))
    lst = file["Time"][peaks]
    gap = 0
    for i in range(len(lst)-1):
        sub = lst.iloc[i+1] - lst.iloc[i] - 1
        gap = sub + gap
    feat2.append(gap)
feat2
```

```python
feat3 = []
for i in entries:
    temp = raw_path + i
    file = pd.read_csv(temp, index_col=0).reset_index()
    peaks, _ = find_peaks(file['2->1Bytes'], height=np.mean(file['2->1Bytes']))
    feat3.append(len(peaks))
feat3
```

# Cobine feats

```python
feats = ["feat1","feat2","feat3"]
data_tuples = list(zip(feat1,feat2,feat3,lab))
tab = pd.DataFrame(data_tuples, columns=['feat1','feat2','feat3','lab'])
tab["feat1"] = tab["feat1"].fillna(1)
y = np.array(tab["lab"])
x = np.array(tab[feats])
#X_train, X_test, y_train, y_test = model_selection.train_test_split(x, y, train_size=0.8,test_size=0.2, shuffle=1)
```

```python
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
acc = []
for i in range(50):
    X_train, X_test, y_train, y_test = model_selection.train_test_split(x, y, train_size=0.8,test_size=0.2, shuffle=1)
    clf = svm.SVC(gamma = "scale")
    clf.fit(X_train, y_train)
    pred = clf.predict(X_train)
    train_accuracy = accuracy_score(pred, y_train)
    #print(train_accuracy)
    pred = clf.predict(X_test)
    test_accuracy = accuracy_score(pred, y_test)
    #print(test_accuracy)
    acc.append(test_accuracy)
np.mean(acc)
```
