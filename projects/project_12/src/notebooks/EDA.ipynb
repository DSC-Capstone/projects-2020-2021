```python
import pandas as pd 
import matplotlib.pyplot as plt
import datetime
import numpy as np
import os
import sys
import seaborn as sns


import pylab as pl
```

```python
def features(df):
    df['total_bytes'] = df.apply(lambda x: x['1->2Bytes'] + x['2->1Bytes'], axis=1)
    df['total_packets'] = df.apply(lambda x: x['1->2Pkts'] + x['1->2Pkts'], axis=1)
    df['index'] = df.index
```

```python
fp = '/teams/DSC180A_FA20_A00/b05vpnxray/group2_data/data'
filelist = os.listdir(fp)
```

```python
def label_identifyer(x):
    if 'netflix' in x:
        return 'Netflix'
    if 'prime' in x:
        return 'Amazon Prime'
    if 'youtube-live'in x:
        return 'Youtube Live'
    if 'youtube' in x and "live" not in x:
        return 'Youtube'
    if 'twitch' in x:
        return "Twitch"
    else:
        return 'Other'
```

```python
def table_output_col(df):
    df['output'] = df['Input'].apply(label_identifyer)
    return df
```

```python
df = pd.DataFrame({"Input": filelist})
df = table_output_col(df)
```
