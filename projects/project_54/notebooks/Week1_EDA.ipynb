```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("train_labels.csv")
df.head()
```

```python
df.shape[0]
```

```python
len(df.scientific_name.unique())
```

```python
country = list(df.country.unique())
df.country.unique()
```

```python
len(country)
```

```python
country_classes = {}
for i in range(df.shape[0]):
    co = df.loc[i]['country']
    cl = df.loc[i]['scientific_name']
    if co not in country_classes:
        country_classes[co] = set()
    country_classes[co].add(cl)
# country_classes
```

```python
countries = list(country_classes.keys())
counts_classes_country =  [len(i) for i in country_classes.values()]
# counts_classes_country
```

```python
# counts_classes_country
```

```python
plt.figure(figsize=(40,20))
# plt.style.use('ggplot')

x_pos = [i for i, _ in enumerate(countries)]

plt.barh(x_pos, counts_classes_country)
plt.ylabel("Country")
plt.xlabel("Number of Classes")
plt.title("Number of Snake Classes Across Different Countries")

plt.yticks(x_pos, countries)

plt.show()
```

```python
df.continent.unique()
```

```python
len(df.continent.unique())
```

```python
continents_classes = {}
for j in range(df.shape[0]):
    co = df.loc[j]['continent']
    cl = df.loc[j]['scientific_name']
    if co not in continents_classes:
        continents_classes[co] = set()
    continents_classes[co].add(cl)
# country_classes
```

```python
# continents_classes
```

```python
continents = list(continents_classes.keys())
counts_classes_continent =  [len(j) for j in continents_classes.values()]
```

```python
plt.figure(figsize=(20,10))
plt.style.use('ggplot')

x_pos = [i for i, _ in enumerate(continents)]

plt.barh(x_pos, counts_classes_continent)
plt.ylabel("Continent")
plt.xlabel("Number of Classes")
plt.title("Number of Snake Classes Across Different Continents")

plt.yticks(x_pos, continents)

plt.show()
```
